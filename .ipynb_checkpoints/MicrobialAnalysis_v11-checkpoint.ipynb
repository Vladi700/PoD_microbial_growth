```python
def thin_discard(chain,discard=0,thin=1):
    #discard
    print(f"Applying discard {discard} and thinning {thin}")
    temp=chain[discard:,:]
    #thinning
    thin_mask=np.arange(0,temp.shape[0],thin)
    temp=temp[thin_mask,:]
    return temp
```

#### Load all flat_chains in flat_chain_list (un-thinned and un-discarded)

```python
#the post_chain files contain the already flattened chains (no need to multiply #rows by the #walkers, it was already done)
flat_chain_list=[]



for filename in files:
    file_path=os.path.join("./modchains",filename)
    with open(file_path,"r") as file: comment = file.readlines()[0]
    values = [int(value) for value in re.findall('[0-9]+', comment)]
    n_steps = values[0]
    n_walker = values[1]
    n_params = values[2]
    print("Loading file:",filename, "with parameters:",values)

    loaded_data = np.loadtxt(file_path)
    print(loaded_data[1])
    print(np.shape(loaded_data))
    print(n_steps*n_walker,n_params)
    #flat_samples=loaded_data.reshape(n_steps*n_walker,n_params)
    #flat_samples = loaded_data.reshape(loaded_data.shape[0] // n_params, n_params)
    flat_samples=loaded_data
    flat_chain_list.append(flat_samples)


print(f"Loaded {len(flat_chain_list)} elements")
for i,element in enumerate(flat_chain_list):
    print (f"{i+1}-esim element, of size {element.nbytes/(1024**2)} MB with shape: ({element.shape})")
```

## Corner plots

```python
def corner_autorange(chains, fig, excess=1/50):
    axes = fig.get_axes()
    flatsamples = chains
    
    for i in np.arange(0,3):
        x_position = axes[i].get_position().x0
        xs = flatsamples[:,i]
        #need to exclude outliers
        #x_min = np.mean(xs)-5*np.std(xs)
        #x_max = np.mean(xs)+5*np.std(xs)
        x_min,x_max=min(xs),max(xs)
        
        for ax in axes:
            if ax.get_position().x0 == x_position:
                ax.set_xlim(x_min - excess*x_min, x_max + excess*x_max)
            
```

### Apply thinning and discard for the purpose of corner plots (now work with temp_flat_chain_list)

```python
#apply thinning and discard
discard=400
thin=25
discard_thin_list=[(discard,thin),(discard,thin),
                   (discard,thin),(discard,thin),
                   (discard,thin),(discard,thin),
                   (discard,thin),(discard,thin),(discard,thin)]

temp_flat_chain_list=[thin_discard(flat_chain,td[0],td[1]) for flat_chain,td in zip(flat_chain_list,discard_thin_list)]
```

### Apply the transformation to every sugar in temp_flat_chain_list

```python
#APPLY THE CHAIN TRANSFORMATION ONLY ONCE, TO APPLY IT AGAIN RELOAD UNTRANSFORMED CHAIN BEFORE
starting_points_list=[]

for i,flat_samples in enumerate(temp_flat_chain_list):
    samplesdf=pd.DataFrame(flat_samples)
    samplesdf.drop(columns=7,inplace=True)#get rid temporarily of log likelihood column
    #initialize old parameters
    u,v,omega2,alpha2,beta2,alpha1,scale1=samplesdf.iloc[0,:7]
    #find initial values for alpha and f
    mean_f=alpha2/(alpha2+beta2)
    std_f=np.sqrt(alpha2*beta2/((alpha2+beta2)*(alpha2+beta2+1)**2))
    mean_alpha=alpha1*scale1
    std_alpha=np.sqrt(alpha1*scale1**2)
    print([u,v,omega2,mean_f,std_f,mean_alpha,std_alpha],"\n")
    

    
    starting_points_list.append([u,v,omega2,mean_f,std_f,mean_alpha,std_alpha])
    
    #f follows a beta distribution
    samplesdf["mean_f"]=samplesdf[3]/(samplesdf[3]+samplesdf[4])
    samplesdf["std_f"]=(samplesdf[3]*samplesdf[4])/((samplesdf[3]+samplesdf[4]+1)*(samplesdf[3]+samplesdf[4])**2)
    samplesdf["std_f"]=np.sqrt(samplesdf["std_f"])

    #alpha follows a gamma distribution
    samplesdf["mean_alpha"]=samplesdf[5]*samplesdf[6] #mean_alpha=alpha*scale
    samplesdf["std_alpha"]=samplesdf[5]*(samplesdf[6])**2#mean_alpha=alpha*(scale)**2
    samplesdf["std_alpha"]=np.sqrt(samplesdf["std_alpha"])

    samplesdf.drop(columns=[3,4,5,6],inplace=True)

    temp_flat_chain_list[i]=samplesdf.to_numpy()#temp chain list don't have the likelihood column
```

```python
#temp_flat_chain_list doesn't have the likelihood column
for starting_points,flat_samples,sugar in zip(starting_points_list,temp_flat_chain_list,sugar_list):
    labels=["u","v","omega2","mean_f","std_f","mean_alpha","std_alpha"]
    param_results=np.around(np.percentile(flat_samples, q=50, axis=0),decimals=5)
    print(f"Corner for {sugar} with starting points after burn-in: \n {starting_points}")
    print(f"arrival points: \n {list(param_results)}")
    fig = corner.corner(flat_samples[:,:],labels=labels,truths=starting_points,show_titles=True,divergences=True)
    
    if sugar=="Lac":#put a non existing sugar to disable the usage of the function
        corner_autorange(flat_samples,fig,excess=1/50)
        plt.savefig(f"emcee_model2_real_7params_{sugar}.svg",format="svg",bbox_inches="tight")
        plt.show()
        plt.close()
    else: 
        plt.savefig(f"emcee_model2_real_7params_{sugar}.svg",format="svg",bbox_inches="tight")
        plt.show()
```

#### Gamma and h computed as h=median(u/v) and gamma=median(omega2/mean_alpha) 

```python
#MEDIAN VALUE
all_sugar_results_new = []#list to store all sugars results

for starting_points,flat_samples,sugar in zip(starting_points_list,temp_flat_chain_list,sugar_list):
    
    labels_new=["u","v","omega2","mean_f","std_f","mean_alpha","std_alpha","h","gamma"]
#extract medians of in particula u,v omega2,mean_alpha
    param_results_new=np.around(np.percentile(flat_samples, q=50, axis=0),decimals=5)
    #print(param_results,"\n")
#extract median of h from the chain of u/v
    param_results_h=np.around(np.percentile(flat_samples[:,0]/flat_samples[:,1], q=50, axis=0),decimals=5)
    #print(param_results_h,"\n")
#extract median of gamma from the chain of omega2/mean_alpha
    param_results_gamma=np.around(np.percentile(flat_samples[:,2]/flat_samples[:,5], q=50, axis=0),decimals=5)
    #print(param_results_gamma,"\n")
    array_h_gamma=np.array([param_results_h,param_results_gamma])
    param_results_tot=np.append(param_results_new,array_h_gamma)
    #print(param_results_tot,"\n\n")
    
    # Create a list to store parameter tuples for iesim sugar results:
    sugar_results_new=[]
    
    for i in range(len(labels_new)):

        parameters_value_new = param_results_tot[i]
        #print(parameters_value,"\n")

        sugar_results_new.append(parameters_value_new)
    all_sugar_results_new.append(sugar_results_new)

# Create a DataFrame from the results:
columns_new = [ 'u', 'v', 'omega2', 'mean_f', 'std_f', 'mean_alpha', 'std_alpha',"h","gamma"]
results_df_new = pd.DataFrame(all_sugar_results_new, columns=columns_new,index=sugar_list)
results_df_new.sort_index(inplace=True)

display(results_df_new[["u","v","omega2","mean_alpha","h","gamma"]])
```

### Credibility intervals (using discard-thinned chains used for corner plot as well)

```python
#CREDIBILITY INTERVALS

all_sugar_bounds = [] #list to store all sugars results

for starting_points,flat_samples,sugar in zip(starting_points_list,temp_flat_chain_list,sugar_list):
    labels_new=["u","v","omega2","mean_f","std_f","mean_alpha","std_alpha","h","gamma"]

    param_results_lower_bounds=np.around(np.percentile(flat_samples, q=2.5, axis=0),decimals=5)#obtain the lower bound for 95% credib interval
    param_results_upper_bounds=np.around(np.percentile(flat_samples, q=97.5, axis=0),decimals=5)#obtain the upper bound for 95% credib interval
    
    param_results_lower_bounds_h=np.around(np.percentile(flat_samples[:,0]/flat_samples[:,1], q=2.5, axis=0),decimals=5)
    param_results_upper_bounds_h=np.around(np.percentile(flat_samples[:,0]/flat_samples[:,1], q=97.5, axis=0),decimals=5)

    param_results_lower_bounds_gamma=np.around(np.percentile(flat_samples[:,2]/flat_samples[:,5], q=2.5, axis=0),decimals=5)
    param_results_upper_bounds_gamma=np.around(np.percentile(flat_samples[:,2]/flat_samples[:,5], q=97.5, axis=0),decimals=5)

    array_lower=np.array([param_results_lower_bounds_h,param_results_lower_bounds_gamma])
    array_upper=np.array([param_results_upper_bounds_h,param_results_upper_bounds_gamma])
    param_lower_tot=np.append(param_results_lower_bounds,array_lower)
    param_upper_tot=np.append(param_results_upper_bounds,array_upper)
    
    # Create tuples for each parameter with credible intervals
   
    sugar_bounds_results=[]
    
    for i in range(len(labels_new)):

        parameters_tuple = (
             param_lower_tot[i],
            param_upper_tot[i],
        )

        sugar_bounds_results.append(parameters_tuple)
    all_sugar_bounds.append(sugar_bounds_results)



temp_flat_chain_list=None # free up memory!

# Create a DataFrame containing results of lower and upper bounds for 95% credibility interval:

results_bounds_df = pd.DataFrame(all_sugar_bounds, columns=columns_new)

results_bounds_df.sort_index(inplace=True)

#alternative: name rows with sugar_list
results_bounds_df.index=sugar_list
```

## Simulating the system to evaluate the fit results

```python
def load_data(fileName,sugar_name="Pyr",out=False): #results_df contains best estimate for un-transformed parameters (beta2,alpha2,alpha1,scale1)
    
    #Loading the file
    path_to_directory = 'data2/'
    df = pd.read_csv(f'{path_to_directory + fileName}', index_col=False,usecols=["generationtime","growth_rate","division_ratio","lineage_ID","generation","length_birth","source"])
    df = df.dropna()
    df=df[df["source"] == sugar_name]
    
    
    # Getting the initial mass
    unique_lineages = df["lineage_ID"].unique()#order from 0 to 99 lineages
    #print(unique_lineages.shape[0])
    initial_mass = np.empty(len(unique_lineages))#create a void array with length equal to the number of unique lineages
    temp_mask=np.ones(len(unique_lineages))
    
    for j,i in enumerate(unique_lineages):
        lineage_mask= (df['lineage_ID'] == i)
        temp=df.loc[lineage_mask]
        
        temp=temp.loc[temp["generation"]==0]#if first mass is not there this will be empty
        if temp.empty:
            df=df[df["lineage_ID"]!=i]#drops all the rows of that lineage 
            print("No initial mass for lineage",i)        
        else: 
            initial_mass[j]=temp["length_birth"].iloc[0]
    
    
    #remove lineages that have any generation time==0
    null_lineages=df[df["generationtime"]==0]["lineage_ID"]
    if out==True: print("Null_lineages:",null_lineages)
    for null_lineage in null_lineages:
        df=df[df["lineage_ID"]!=null_lineage]
    return df,initial_mass,unique_lineages

```

```python
import numpy as np
from scipy.stats import gamma,beta
from scipy.optimize import root_scalar

import matplotlib.pyplot as plt
import pandas as pd

#gamma:
def f_alpha_i(params,alpha):
    tau, k, alpha_S, alpha2, beta2, alpha1, scale1 = params
    return gamma.pdf(alpha,a=alpha1,scale=scale1)

#beta:
def f_f_i(params,f_i):
    tau, k, alpha_S, alpha2, beta2, alpha1, scale1 = params
    return beta.pdf(f_i,a=alpha2,b=beta2)

''' Logarithm of likelihood of all measurement'''

def log_likelihood(params, initial_mass, lineages, t, alpha, f_i):
    tau, k, alpha_S, alpha2, beta2, alpha1, scale1 = params
    masses_before_division = np.array([])

    for j, i in enumerate(np.unique(lineages)):
        filtered_times = t[lineages == i]
        filtered_alphas = alpha[lineages == i]
        filtered_fs = f_i[lineages == i]
        masses_before_division_per_lineage = evolve_mass(params, filtered_times, filtered_alphas, filtered_fs, initial_mass[j])
        masses_before_division = np.concatenate((masses_before_division, masses_before_division_per_lineage))

    first = p(t, tau, k, alpha_S, masses_before_division, alpha, mode="generation")
    second = f_alpha_i(params, alpha)
    third = f_f_i(params, f_i)

    first = np.where(first == 0, 1e-300, first)
    second = np.where(second == 0, 1e-300, second)
    third = np.where(third == 0, 1e-300, third)

    res = np.sum(np.log(first) + np.log(second) + np.log(third))
    return res

'''Hazard function (probability of having a division with certain parameters)'''

#in generation data m_D changes every time wherease in fit data it is a constant=initial_mass

def protein(t,m_D, alpha):
    return m_D *( np.exp(alpha*t) - 1)


"""def h(t, tau, k, alpha_S, m_D, alpha, mode="generation"):
    p = protein(t, m_D, alpha)
    dpdt = m_D * alpha * np.exp(alpha * t)

    # Core hazard function in terms of protein
    hazard_core = (alpha_S * k / tau) * ((p / tau)**(k - 1)) / (1 + (p / tau)**k)
    h_t = hazard_core * dpdt

    if mode == "generation":
        return h_t
    elif mode == "inference":
        return h_t"""

    
"""def s(t, tau, k, alpha_S, m_D, alpha, mode="generation"):
    p = protein(t, m_D, alpha)
    S_p = 1 / (1 + (p / tau)**k)**alpha_S
    if mode == "generation":
        # If you want to introduce a t0 cutoff as before, do it here
        return S_p
    elif mode == "inference":
        return S_p"""

def h(t, tau, k, alpha_S, m_D, alpha, mode="generation"):
    p = protein(t, m_D, alpha)
    dpdt = m_D * alpha * np.exp(alpha * t)
    hazard_core = (alpha_S * k / tau) * ((p / tau) ** (k - 1)) / (1 + (p / tau) ** k)
    return hazard_core * dpdt


def s(t, tau, k, alpha_S, m_D, alpha, mode="generation"):
    p = protein(t, m_D, alpha)
    return 1 / (1 + (p / tau) ** k) ** alpha_S
   

############################################################################
#used only for generating synthetic data:
'''Survival function used for the sampling (and only there)'''

def s_root(t, tau, k, alpha_S, m_D, alpha):
    return s(t, tau, k, alpha_S, m_D, alpha)


def t0_root(tau, m_D, alpha):
    res = (tau / m_D) + 1
    return np.log(res) / alpha

''' Logarithm of prior function '''
def log_prior(params):
    tau, k, alpha_S, alpha2, beta2, alpha1, scale1 = params
    if alpha2 > 0 and beta2 > 0 and alpha1 > 0 and scale1 > 0 and alpha_S > 0 and tau > 0 and tau < 20 and k > tau and k < 20:
        return 0.0
    else:
        return -np.inf


def log_probability(params, initial_mass, lineages, t, alpha, f_i):
    lp = log_prior(params)
    if not np.isfinite(lp):
        return -np.inf
    res = lp + log_likelihood(params, initial_mass, lineages, t, alpha, f_i)
    return res

############################################################################

'''Distribution of times of divisions'''
#is the one that we have called first in log_loikelihhod function
def p(t, tau, k, alpha_S, m_D, alpha, mode="generation"):
    return h(t, tau, k, alpha_S, m_D, alpha, mode) * s(t, tau, k, alpha_S, m_D, alpha, mode)



'''Mass at division'''
def m(t, m_D, alpha):
    return m_D * np.exp(alpha * t)


def t0(params, m_D, alpha):
    tau, k, alpha_S, alpha2, beta2, alpha1, scale1 = params
    res = (tau / m_D) + 1
    return np.log(res) / alpha


def evolve_mass(params, filtered_times, filtered_alphas, filtered_fs, initial_mass):
    masses = np.empty(filtered_times.shape[0])
    mass = initial_mass
    for i in range(masses.shape[0]):
        masses[i] = mass
        mass = m(filtered_times[i], mass, filtered_alphas[i]) * filtered_fs[i]
    return masses



''' Chain plotter '''
#To plot emcee outputs(evolution of parameter values found by random walkers)

def chainplot(chains, interval, parameter, n_plotted=6, random_pick=False, discard=200, thin=25, legend=False):
    
    n_walkers = chains.shape[1]
    
    if random_pick == True:
        plotted_walkers = np.random.choice(np.arange(n_walkers), n_plotted, replace=False)
    else:
        plotted_walkers = np.arange(0, n_plotted)

    return plotted_walkers



#### example, steal this box
#interval = np.arange(3500, chain.shape[0])
#model_2.chainplot(sampler=sampler,       # ensemble sampler from emcee, after it is run
#                  interval=interval,     # which interval of steps is plotted  
#                  parameter = 7,         # which parameter evolution is plotted (counting from 0)
#                  n_plotted = 5,         # default 6 --- number of plotted walkers
#                  random_pick = True,    # default False --- if true pick the walkers randomly, if false the first n_plotted ones 
#                  discard=100,           # default 100 --- parameter for get_chain method, discard the first entries
#                  thin=1,                # default 1 --- pick a sample every n (n=1 -> pick all samples)
#                  legend=True            # default False --- if true, plot legend
#                 )
####


''' Chain sampler'''

def chainsampler (sampler, params, n_samples, n_walkers=1, discard=0, make_hist=False, save_data = False):
    chains = sampler.get_chain(discard = 0)
    chain_length = chains.shape[0]
    
    
    sampled_walker_ids = np.random.choice(np.arange(chains.shape[1]), n_walkers, replace=False)
    chains_picked = chains[:,sampled_walker_ids,:]
    steps_to_pick = np.linspace(0, chain_length, n_samples, dtype=int)
    steps_to_pick = np.delete(steps_to_pick, -1)
    
    if make_hist==True:
        plt.hist(chains_picked[steps_to_pick])
        plt.show
    
    print('Picked walkers id: ', str(sampled_walker_ids))
    
    if save_data==True:
        df = pd.DataFrame(chains_picked[steps_to_pick])
        df.to_csv('chains_ids.csv', index=False)  
        
    return(chains_picked[steps_to_pick,:,:])  

'''Generate synthetic data from an initial mass'''

def evolve(initial_mass=None, masses_list=None, n_cells=1, n_times=10, params=None, t_max=30.0):
  '''Generates synthetic data of n_times divisions for n_cells lineages, all with the same initial mass
    The params are u, v, omega2, alpha1, scale1, alpha2, beta2
    t_max is the max time of division used as a treshold for the root finding algorithm"
  '''
  u, v, omega2,alpha2, beta2,alpha1, scale1 = params
  
  data = {'lineage_ID': [], 'generation_N':[], 'mass_birth':[], 'time_division':[], 'alpha':[], 'f':[]}

  for j in range(n_cells): #For each lineage of cells

    #At the beginning of each lineage the initial mass is the same of the progenitor
    mass = masses_list[j] if initial_mass==None else initial_mass
      
    for i in range(n_times):  #Follow the divisions
      data['generation_N'].append(i)
      data['lineage_ID'].append(j)
      #Generate alpha and f from the respective distributions and store the values in the list
      alpha_D = gamma.rvs(a=alpha1,scale=scale1)  #DO NOT set size=1 because it returns an array instead of a number
      data['alpha'].append(alpha_D)

      f_D=beta.rvs(a=alpha2,b=beta2)
      data['f'].append(f_D)

      #generate random division time
      val = np.random.uniform()
      f_to_root = lambda t, val: 1 - s_root(t,u,v,omega2,mass,alpha_D) - val 
      t_0 = t0_root(u, mass, alpha_D)

      try:
            t_D = root_scalar(f_to_root, bracket=(t_0, t_max),  args=(val,), method='brentq').root 
      except ValueError: 
            t_D = t_max
      data['time_division'].append(t_D)

      #Get mass before division (obtained from previous cycle mass )
      data['mass_birth'].append(mass)
      m_before_D = m(t_D, mass, alpha_D) #where alpha is sampled from Gamma distribution
      #Get mass after division (sample f randomly from Beta distr)
      m_after_D = f_D*m_before_D    #<-where f is sampled from Beta distribution
      mass = m_after_D
    
    # Save masses and times in a global list and use the last masses as starting point for the next step

  return pd.DataFrame(data)

'''Survival function used for the sampling (and only there)'''

# Modified s_root to use the new 's' function and its parameters
def s_root(t, tau, k, alpha_S, m_D, alpha):
    return s(t, tau, k, alpha_S, m_D, alpha, mode="generation")

# Modified t0_root to use 'tau' instead of 'u'
def t0_root(tau, m_D, alpha):
    res = (tau / m_D) + 1
    return np.log(res) / alpha
```
